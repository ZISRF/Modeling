```python
# Simulation

g = 9.81                # m / s^2
k = 9e9
e0 = 8.85e-12

Q = 1e-2                # K
S = 1                   # m^2
D = 1                   # m
e = 1

m = 9.1e-32             # kg
q = 1.6e-19             # K
x = 0.5                 # m
y = 0                   # m
v_x = 0                 # m / s
v_y = 100               # m / s
t = 0

T = []
X = []
Y = []
Vx = []
Vy = []
Ax = []
Ay = []

dt = 1e-15
while x > 0 and x < D:
    t += dt
    T.append(t)

    a_x = q * Q / (e * e0 * S * m)
    a_y = -g
    Ax.append(a_x)
    Ay.append(a_y)

    x += v_x * dt + a_x * dt * dt / 2
    y += v_y * dt + a_y * dt * dt / 2
    X.append(x)
    Y.append(y)

    v_x += a_x * dt
    v_y += a_y * dt
    Vx.append(v_x)
    Vy.append(v_y)
```

```python
# Plot X(Y)

import matplotlib.pyplot as plt
import numpy as np

plt.xlabel("X, m")
plt.ylabel("Y, m")
plt.plot(X, Y)

plt.plot([0, 0], [np.min(Y), np.max(Y)], color='red', linewidth=5)
plt.plot([D, D], [np.min(Y), np.max(Y)], color='red', linewidth=5)
```
